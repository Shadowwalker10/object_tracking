```python
import cv2
import numpy as np
from tkinter import *
from tkinter.ttk import *
```

```python
##OpenCV has 8 different tracking algorithms
#trackers = ['BOOSTING', 'MIL','KCF', 'TLD', 'MEDIANFLOW', 'GOTURN', 'MOSSE', 'CSRT']

root = Tk()
root.geometry("300x300")
root.title("Tracking Methods")

def selected():
    vals = main.get()
    return vals


main = StringVar(value = "2")

trackers = ['BOOSTING', 'MIL','KCF', 'TLD', 'MEDIANFLOW', 'GOTURN', 'MOSSE', 'CSRT']

label = Label(text = "Select the tracker: ")
label.pack(fill = 'x',padx = 5,pady = 5)

for tracks in trackers:
    r = Radiobutton(root, text = tracks,value = tracks,variable = main)
    r.pack(fill = 'x',padx = 5,pady = 5)

root.mainloop()

tracker_type = selected()



#For new usage of tracker, we need opencv-contrib-python

if tracker_type == 'BOOSTING':
    tracker = cv2.legacy.TrackerBoosting_create()
if tracker_type == 'MIL':
    tracker = cv2.TrackerMIL_create() 
if tracker_type == 'KCF':
    tracker = cv2.TrackerKCF_create() 
if tracker_type == 'TLD':
    tracker = cv2.legacy.TrackerTLD_create() 
if tracker_type == 'MEDIANFLOW':
    tracker = cv2.legacy.TrackerMedianFlow_create() 
if tracker_type == 'GOTURN':
    tracker = cv2.TrackerGOTURN_create()
if tracker_type == 'MOSSE':
    tracker = cv2.legacy.TrackerMOSSE_create()
if tracker_type == "CSRT":
    tracker = cv2.TrackerCSRT_create()
    


cap = cv2.VideoCapture(0)
width = 640
height = 480
cap.set(3,width)
cap.set(4,height)


#Allowing users to select the frame to track

success,img = cap.read()
img = cv2.flip(img,1)
bbox = cv2.selectROI("Track",img,False)
#This bbox is a tuple


#Initalize the tracker
tracker.init(img,bbox)

def drawbox(img,bbox):
    x,y,w,h = int(bbox[0]),int(bbox[1]),int(bbox[2]),int(bbox[3])
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.putText(img,"Tracking",(10,80),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,0,255),2)
    

while True:
    timer = cv2.getTickCount()
    success,img = cap.read()
    img = cv2.flip(img,1)
    fps = round(cv2.getTickFrequency()/(cv2.getTickCount()-timer),2)
    
    success,bbox = tracker.update(img)
    if success:
        drawbox(img,bbox)
    
    else:
        cv2.putText(img,"Lost the object",(10,80),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,0,0),2)
    
    
    
    cv2.putText(img,"FPS: "+str(fps),(10,50),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,0,0),2)
    
    cv2.imshow("Track",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

**Note don't forget to press enter after you select the area to track**
